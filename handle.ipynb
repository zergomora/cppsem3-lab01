```python
import pandas as pd 

df = pd.read_csv("second.csv", delimiter='\t')

df
```

```python
import matplotlib.pyplot as plt 

plt.bar(df["Target"], df["Standart"])

plt.title('Standart Dice')
plt.ylabel(r'Вероятность попадания $\omega$')
plt.xlabel('Значение X')

plt.xlim([1,100])
plt.show()
```

```python
plt.bar(df["Target"], df["Three"])

plt.title('Three Dice')
plt.ylabel(r'Вероятность попадания $\omega$')
plt.xlabel('Значение X')

plt.xlim([0,20])
plt.show()
```

```python
plt.bar(df["Target"], df["Penalty"], label="Penalty")
plt.bar(df["Target"], df["Bonus"], label="Bonus")

plt.title('Penalty&Bonus Dice')
plt.ylabel(r'Вероятность попадания $\omega$')
plt.xlabel('Значение X')

plt.xlim([1,100])
plt.legend()
plt.show()
```
